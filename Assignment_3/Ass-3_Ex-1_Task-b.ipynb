# Theoretical Background

## Sampling and Frequency Shifting
Sampling a continuous-time signal at a discrete rate, $f_s$, replicates its spectrum at intervals of $f_s$. If $f_s$ is less than twice the highest frequency component of the signal, overlapping spectra can result, leading to aliasing.

## Aliasing
Aliasing is a distortion that occurs when higher frequency components of a signal are misinterpreted as lower frequencies during the sampling process. It happens because the shifted spectra overlap, obscuring the true frequency content of the signal.


# Mathematical Development

Given the original frequencies $f_1 = 4$ kHz and $f_2 = 6$ kHz, the sampling frequency $f_s = 10$ kHz causes the spectrum to repeat every $f_s$ Hz. The sampling frequency used here is insufficient to meet the Nyquist criterion, which requires at least $12$ kHz to avoid aliasing. This exercise will analyze how these insufficiencies affect the spectrum, particularly looking for overlaps within the range of $-f_s$ to $+f_s$.

# Python Code Implementation
The following Python code calculates and visualizes the shifted spectra and the resulting spectrum of the sampled signal $x[n]$. The code helps to demonstrate visually how aliasing manifests due to spectral overlap when the sampling frequency is below the Nyquist rate.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sampling frequency and original signal frequencies
fs = 10000  # Sampling frequency in Hz
original_frequencies = np.array([-6000, -4000, 4000, 6000])  # Original frequencies in Hz

# Generate shifted frequencies for -fs, 0, +fs
frequencies_minus_fs = original_frequencies - fs
frequencies_plus_fs = original_frequencies + fs

# Amplitudes for plotting clarity
amplitudes_original = np.ones_like(original_frequencies) * 1.0  # Amplitude for original frequencies
amplitudes_minus_fs = np.ones_like(frequencies_minus_fs) * 0.5  # Reduced amplitude for -fs shifted frequencies
amplitudes_plus_fs = np.ones_like(frequencies_plus_fs) * 0.5  # Reduced amplitude for +fs shifted frequencies

# Plotting the original and shifted spectra
plt.figure(figsize=(12, 6))
plt.stem(original_frequencies, amplitudes_original, linefmt='r-', markerfmt='ro', basefmt=" ", label='Original Frequencies')
plt.stem(frequencies_minus_fs, amplitudes_minus_fs, linefmt='b-', markerfmt='bo', basefmt=" ", label='Shifted by -fs')
plt.stem(frequencies_plus_fs, amplitudes_plus_fs, linefmt='g-', markerfmt='go', basefmt=" ", label='Shifted by +fs')

# Adjust plot settings
plt.title('Shifted Spectra and Resulting Spectrum of $x[n]$ Across Extended Range')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim(-2 * fs, 2 * fs)  # Set x-axis limits from -2fs to +2fs
plt.legend()
plt.grid(True)
plt.savefig('fig/ex1_b_plot.png')
plt.show()
```

Keeping amplitudes to avoid misinterpretation:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sampling frequency and original signal frequencies
fs = 10000  # Sampling frequency in Hz
original_frequencies = np.array([-6000, -4000, 4000, 6000])  # Original frequencies in Hz

# Generate shifted frequencies for -fs, 0, +fs
frequencies_minus_fs = original_frequencies - fs
frequencies_plus_fs = original_frequencies + fs

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(12, 6))

# Plot original frequencies with larger markers first
markerline, stemlines, baseline = ax.stem(original_frequencies, np.ones_like(original_frequencies), linefmt='r-', markerfmt='ro', basefmt=" ", label='Original Frequencies')
plt.setp(markerline, 'markersize', 10)  # Adjust marker size for original frequencies

# Plot shifted by -fs with smaller markers
markerline, stemlines, baseline = ax.stem(frequencies_minus_fs, np.ones_like(frequencies_minus_fs), linefmt='b--', markerfmt='bo', basefmt=" ", label='Shifted by -fs')
plt.setp(markerline, 'markersize', 6)  # Adjust marker size for shifted frequencies

# Plot shifted by +fs with smaller markers
markerline, stemlines, baseline = ax.stem(frequencies_plus_fs, np.ones_like(frequencies_plus_fs), linefmt='g--', markerfmt='go', basefmt=" ", label='Shifted by +fs')
plt.setp(markerline, 'markersize', 6)  # Adjust marker size for shifted frequencies

# Adjust plot settings
ax.set_title('Shifted Spectra and Resulting Spectrum of $x[n]$ Across Extended Range')
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Amplitude')
ax.set_xlim(-2 * fs, 2 * fs)  # Set x-axis limits from -2fs to +2fs
ax.legend()
ax.grid(True)

# Show the plot
plt.show()
```

# Conclusion

This visualization clarifies the effects of undersampling on the frequency spectrum. It explicitly shows how overlapping spectral components due to a sampling frequency of $10$ kHz, which is not sufficient to prevent aliasing, can lead to significant distortion in the sampled signal. Properly choosing a sampling frequency that complies with the Nyquist criterion is essential for maintaining the integrity of the original signal in digital representations.
