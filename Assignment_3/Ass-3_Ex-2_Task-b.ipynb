# Introduction
In Exercise 2, Task (b), we explore the effects of sampling the analog signal \(x(t)\) at 8 kHz. The task involves calculating and visualizing the spectrum of the resultant discrete-time signal \(x[n]\) and identifying the baseband, which is crucial for understanding aliasing and spectral replication due to sampling.


# Theoretical Background

## Sampling and Spectrum Replication
When an analog signal is sampled, its spectrum is periodically replicated at intervals equal to the sampling frequency (\(f_s\)). For a sampling rate of 8 kHz, these replications occur every 8 kHz. This replication can lead to overlapping of spectral components, known as aliasing, if parts of the spectrum extend beyond the Nyquist frequency (\(f_{Nyquist} = f_s/2 = 4 kHz\)).


# Python Code for Spectrum Visualization
The following Python code computes and visualizes the spectrum of \(x[n]\) from \(-f_s\) to \(f_s\). It also highlights the baseband to help identify any potential aliasing effects.


```python
import numpy as np
import matplotlib.pyplot as plt

fs = 8000  # Sampling frequency in Hz
f_nyquist = fs / 2  # Nyquist frequency
f = np.linspace(-fs, fs, 8000)  # Frequency range for plotting

# Define the amplitude A for visualization clarity
A = 1

# Define the magnitude and phase functions based on piecewise conditions
def magnitude_spectrum(f):
    return np.where((np.abs(f) >= 1) & (np.abs(f) <= 5), A,
            np.where((f > -1) & (f < 0), -A * f,
            np.where((f >= 0) & (f < 1), A * f, 0)))

# Generate the spectrum for a wide range to include multiple replications
X_f = magnitude_spectrum(f)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(f, X_f, label='Spectrum of $x[n]$', color='blue')
plt.axvline(x=-f_nyquist, color='red', linestyle='--', label='-$f_{Nyquist}$')
plt.axvline(x=f_nyquist, color='green', linestyle='--', label='$f_{Nyquist}$')
plt.title('Spectrum of the Sampled Signal $x[n]$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.savefig('fig/ex2_task_b_spectrum.png')  # Save the figure
plt.show()
```

# Conclusion
The visualization demonstrates how the spectrum of \(x[n]\) is affected by sampling at 8 kHz. Notably, the plot highlights the baseband and the effects of spectral replication, providing insight into potential aliasing and the need for careful filter design in digital signal processing applications.
