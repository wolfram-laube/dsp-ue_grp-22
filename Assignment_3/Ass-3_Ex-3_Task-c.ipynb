# Introduction
In Exercise 3, Task (c), we undertake the qualitative visualization of the spectrum of the analog signal \(x(t)\) after it has been converted from the discrete-time signal \(x[n]\) using a DAC with zero-order hold, up to 20 kHz. This task focuses on identifying the effect of the DAC's sampling rate and zero-order hold on the frequency components of \(x(t)\).


# Theoretical Background

## Digital-to-Analog Conversion with Zero-Order Hold
The zero-order hold (ZOH) reconstruction method in DACs holds each sampled value constant until the next sample, which introduces a sinc-shaped distortion in the frequency domain. This distortion modifies the amplitude of the signal's frequency components.

## Spectrum Analysis
Given the base frequency of \(x[n]\) and the effects of ZOH, the spectrum of \(x(t)\) will exhibit delta pulses at the base frequency and its harmonics, modulated by the sinc function due to the sampling process.


# Python Code for Spectrum Visualization
Below is the Python code to visualize the spectrum of \(x(t)\) qualitatively, displaying delta pulses up to 20 kHz, with arrow heights representing the relative weights influenced by the ZOH effect.


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
fs = 8000  # Sampling frequency in Hz
f0 = 1000  # Base frequency in Hz
T = 1 / fs  # Sampling period

# Generate frequency values and calculate the sinc amplitude for visualization
frequencies = np.arange(-20000, 20001, 1000)  # From -20 kHz to 20 kHz in 1 kHz steps
sinc_amplitudes = np.sinc(frequencies * T)  # Sinc function values for each frequency

# Plot
plt.figure(figsize=(10, 6))
plt.stem(frequencies, sinc_amplitudes, use_line_collection=True)
plt.title('Qualitative Spectrum of $x(t)$ up to 20 kHz')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (relative)')
plt.grid(True)
plt.xlim(-20000, 20000)

# Highlighting the baseband and significant components
for f, amp in zip(frequencies, sinc_amplitudes):
    if abs(f) % fs == f0:
        plt.annotate(f'{f//1000} kHz', (f, amp), textcoords="offset points", xytext=(0,10), ha='center')

plt.savefig('fig/ex3_task_c_spectrum.png')  # Save the figure
plt.show()
```

# Conclusion
The visual representation of the spectrum illustrates the effects of the sampling rate and zero-order hold on the analog signal's frequency components. This qualitative analysis aids in understanding the spectral implications of DAC methods in digital signal processing.
