# Introduction
In this part of Exercise 3, Task (a), we analyze the frequency components of the analog signal \(x(t)\) derived from a discrete-time signal \(x[n]\) through a digital-to-analog converter (DAC) operating at 8 kHz. Our goal is to list all positive frequencies occurring in \(x(t)\) up to 20 kHz.


# Theoretical Background

## Digital-to-Analog Conversion
When a discrete-time signal is converted to an analog signal using a DAC, the original frequency components are replicated around multiples of the sampling frequency. This process can introduce additional frequency components, known as aliasing, depending on the spectral content of the discrete signal and the sampling rate.

## Frequency Analysis
Given the sinusoidal nature of \(x[n]\) and the sampling frequency, we will determine the fundamental and replicated frequencies in the analog signal \(x(t)\).


# Python Code for Frequency Analysis
The following Python code will list and visualize the positive frequencies up to 20 kHz that occur in the analog signal \(x(t)\) after being processed by the DAC. This analysis is based on identifying the fundamental frequency of the discrete signal and observing how it and its replicas appear in the spectrum due to the sampling frequency.


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
fs = 8000  # Sampling frequency in Hz
f0 = 1000  # Base frequency in Hz
max_freq = 20000  # Maximum frequency to list

# Generate the list of frequencies
frequencies = []
for k in range(-3, 4):  # Check around several multiples of fs
    f_plus = k * fs + f0
    f_minus = k * fs - f0
    if 0 < f_plus <= max_freq:
        frequencies.append(f_plus)
    if 0 < f_minus <= max_freq:
        frequencies.append(f_minus)

# Remove duplicates and sort
frequencies = sorted(set(frequencies))

# Output the frequencies
print("Positive frequencies occurring in x(t) up to 20 kHz:")
for f in frequencies:
    print(f"{f} Hz")

# Plotting
plt.figure(figsize=(10, 5))
plt.stem(frequencies, np.ones(len(frequencies)), basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title('Positive Frequencies in $x(t)$ up to 20 kHz')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Presence')
plt.grid(True)
plt.savefig('fig/ex3_task_a_frequencies.png')  # Save the figure
plt.show()
```

# Conclusion
This analysis highlights the positive frequencies that appear in the analog signal \(x(t)\) due to digital-to-analog conversion. Understanding these frequency components is crucial for ensuring accurate signal processing and avoiding issues like aliasing.
