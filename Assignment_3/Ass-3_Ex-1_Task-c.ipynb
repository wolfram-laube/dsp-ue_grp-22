# Introduction
This section addresses Task (c) of Exercise 1, where we examine the visual representation of the continuous signal $x(t)$ and its sampled counterpart $x[n]$. We aim to demonstrate how accurately the sampled signal represents the continuous signal when sampled at a high rate of $100$ kHz, and to show how this relates to the spectrum analysis conducted in the previous task.


# Theoretical Background

## Continuous vs. Sampled Signals
A continuous signal, such as $x(t)$, can be sampled to produce a discrete sequence $x[n]$. When the sampling frequency is sufficiently high relative to the signal's highest frequency component, the sampled signal can closely approximate the original continuous signal.

## Importance of High Sampling Rates
Sampling at a rate significantly higher than twice the maximum frequency component of the signal (as per the Nyquist criterion) ensures that the original signal's information is preserved and that aliasing is avoided. This task visually explores these concepts.


# Python Code Implementation
The following Python code simulates both the continuous signal $x(t)$ and its sampled version $x[n]$, plotted over the same time interval to illustrate the fidelity of the sampling process. The code also saves the plot to be used in the LaTeX document.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal parameters
f1 = 4000  # Frequency of the first sine wave in Hz
f2 = 6000  # Frequency of the second sine wave in Hz
fs = 100000  # Sampling frequency in Hz

# Time arrays for continuous and sampled signals
t = np.linspace(0, 0.002, 1000, endpoint=False)  # Continuous time array from 0 to 2 ms
ts = np.arange(0, 0.002, 1/fs)  # Sampled time array at intervals of Ts = 1/fs

# Signal definitions
x_t = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)  # Continuous signal
x_n = np.sin(2 * np.pi * f1 * ts) + np.sin(2 * np.pi * f2 * ts)  # Sampled signal

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x_t, label='Continuous $x(t)$', linestyle='-', color='blue')
plt.stem(ts, x_n, linefmt='r-', markerfmt='ro', basefmt='r-', label='Sampled $x[n]$')
plt.title('Continuous and Sampled Signal Representation')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Save the figure to a file for inclusion in the LaTeX document
plt.savefig('fig/ex1_c_plot.png')

# Show the plot
plt.show()
```

# FFT Analysis
To further validate the sampling process and the conclusions from Task (b), we'll perform a Fast Fourier Transform (FFT) on the sampled signal $x[n]$. This will allow us to visually confirm the frequency components identified previously and demonstrate the effectiveness of the sampling rate in capturing the signal's true characteristics without aliasing.


# Python Code for FFT Analysis
The following code computes the FFT of the sampled signal $x[n]$ and plots its frequency spectrum. This visual representation will help confirm that the sampling rate of 100 kHz is sufficient to capture all significant frequency components of the signal.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal parameters
f1 = 4000  # Frequency of the first sine wave in Hz
f2 = 6000  # Frequency of the second sine wave in Hz
fs = 100000  # Sampling frequency in Hz

# Generate the sampled time array and the corresponding signal
ts = np.arange(0, 0.002, 1/fs)  # Sampled time array at intervals of Ts = 1/fs
x_n = np.sin(2 * np.pi * f1 * ts) + np.sin(2 * np.pi * f2 * ts)  # Sampled signal

# Compute the FFT and corresponding frequency bins
N = len(ts)  # Number of points in FFT
X_n = np.fft.fft(x_n)  # FFT of sampled signal
frequencies = np.fft.fftfreq(N, 1/fs)  # Frequency bins

# Plotting the FFT
plt.figure(figsize=(10, 6))
plt.stem(frequencies[:N // 2], np.abs(X_n[:N // 2]), linefmt='b-', markerfmt='bo', basefmt='b-', label='FFT of $x[n]$')
plt.title('Frequency Spectrum of the Sampled Signal $x[n]$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, 10000)  # Limit the x-axis for clarity
plt.legend()
plt.grid(True)

# Save the figure to a file for inclusion in the LaTeX document
plt.savefig('fig/ex1_c_fft_plot.png')

# Show the plot
plt.show()
```

# Conclusion
The plot clearly illustrates that the sampled signal $x[n]$, when sampled at $100$ kHz, closely follows the continuous signal $x(t)$. This visual confirmation supports our spectral analysis findings from Task (b), showing that high sampling rates can effectively capture the signal's true characteristics, thereby avoiding aliasing and preserving signal integrity.
